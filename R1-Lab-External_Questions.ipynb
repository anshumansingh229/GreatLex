### Find out the no of male & female employees in the company

*Hint: Use can use a condition to find out the no of 1s and 0s in gender and sum it*

```python
df['Gender']
```

```python
#Seeing the above output looks like Gender has two values 1,2, I am considering 1 -> Male and 2->Female
maleCount = 0
femaleCount = 0
for count in list(df['Gender']):
    if(count == 1):
        maleCount = maleCount + 1
    else:
        femaleCount = femaleCount + 1

print ("Male Count : {}, Female Count: {}".format(maleCount,femaleCount))
```

### Find out the IQR for the column - Age

```python
q1 = df['Age'].quantile(0.25)
print(q1)
```

```python
q3 = df['Age'].quantile(0.75)
print(q3)
```
