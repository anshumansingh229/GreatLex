```python
#Importing Laliga.csv and Printing the head
#Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data
import pandas as pd
#Read the data and stored it in df = Laliga
Laliga = pd.read_csv("C:\\Users\\anshumans\\Downloads\\Laliga.csv")
```

```python
#Question-1
#Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data.
#To check if any entry with - before replacement
print((Laliga == '-').sum())
```

```python
#To replace the existing "-" with 0 and assing it back to the Laliga data frame
Laliga = Laliga.replace("-", 0)
print(Laliga)
```

```python
#To check if any entry with - after replacement
print((Laliga == '-').sum())
```

```python
#Question 2
#Print all the teams which have started playing between 1930-1980.
debutYear = Laliga["Debut"].str.split("-", n = 1, expand = True)
#Added the DebutYear into the Laliga df
Laliga["DebutYear"]= debutYear[0]
#To validate if Debut Year has been added to Laliga
print(Laliga)
```

```python
#Printing the Team which has debut year between 1930-1980 both inclusive
print(Laliga.Team[Laliga['DebutYear'].astype(str).astype(int).between(1930, 1980, inclusive=True)])
```

```python
#Question 3
#Print the list of teams which came Top 5 in terms of points
#Need to convert object type to int before comparing
#Printing the Top 5 team
Laliga.Team[Laliga['Points'].astype(str).astype(int).isin(Laliga['Points'].astype(str).astype(int).nlargest(5))]
```

```python
#Question 4
#Write a function with name “Goal_diff_count” which should return all the teams with their Goal 

def Goal_diff_count(Laliga):
    Laliga['goalDifference'] = [int(a)-int(b) for a,b in zip(list(Laliga['GoalsFor']),list(Laliga['GoalsAgainst']))]
    return Laliga

    
Laliga = Goal_diff_count(Laliga)
#Adding print to validate the Tema and its goal difference
print(Laliga[['Team','goalDifference']])
```

```python
#Printing the top 5 team with largest Goal difference
Laliga.Team[Laliga['goalDifference'].astype(str).astype(int).isin(Laliga['goalDifference'].astype(str).astype(int).nlargest(5))]
```

```python
#Printing the top 5 team with smallest Goal difference
Laliga.Team[Laliga['goalDifference'].astype(str).astype(int).isin(Laliga['goalDifference'].astype(str).astype(int).nsmallest(5))]
 
```

```python
#Question  5
#Create a new column with name “Winning Percent” and append it to the data set

def Win_Perc_count(Laliga):
    winPercentage = []
    for (gamesWon,gamesPlayed) in zip (list(Laliga['GamesWon']) , list(Laliga['GamesPlayed'])):
        try:
            winPercentage.append((float(gamesWon) / float(gamesPlayed)) * 100) 
        except:
            winPercentage.append(0)
    return winPercentage
    
    
Laliga['Winning percentage'] = Win_Perc_count(Laliga)
print(Laliga[['Team','Winning percentage']])
```

```python
#Printing Top 5 Wining prcentage Team
Laliga.Team[Laliga['Winning percentage'].astype(str).astype(float).isin(Laliga['Winning percentage'].astype(str).astype(float).nlargest(5))]
```

```python
#Question 6
#Group teams based on their “Best position” and print the sum of their points for all positions 
print(Laliga['Points'].astype(str).astype(int).groupby(Laliga['BestPosition']).sum())
```
